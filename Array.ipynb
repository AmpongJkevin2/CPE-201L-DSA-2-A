```python
def traverse_array(arr):
    """Function to traverse and print all elements in the array"""
    print("Traversing the array:")
    for i, element in enumerate(arr):
        print(f"Index {i}: {element}")

def reverse(arr: list):
    print("reversing all elements")
    arr.reverse()
    for i, element in enumerate(arr):
        print(f"Index {i}: {element}")

def sort(arr: list):
    print("sort all elements")
    arr.sort()
    for i, element in enumerate(arr):
        print(f"Index {i}: {element}")

def insert(arr: list):
    print("Insert element")
    index, new_element = input("enter the index, value: ").split(" ")
    arr.insert(int(index), int(new_element))

def delete(arr: list):
    print("remove element")
    target_index = input("enter the index: ")
    del arr[int(target_index)]

def modify(arr: list):
    print("modify element")
    target_index = input("enter the index: ")
    new_value = input("enter new value: ")
    arr[int(target_index)] = int(new_value)


def main():
    # Initialize an array
    arr = [10, 20, 30, 40, 50]
    print("Original array:", arr)


    while True:
        print("\nMenu:")
        print("1. Traverse array")
        print("2. Reverse array")
        print("3. sort array")
        print("4. insert array")
        print("5. delete arrat")
        print("6. modify array")
        print("7. Exit")
        choice = input("Enter your choice (1-7): ")

        match choice:
            case '1':
                traverse_array(arr)
            case '2':
                reverse(arr)
            case '3':
                sort(arr)
            case '4':
                insert(arr)
            case '5':
                delete(arr)
            case '6':
                modify(arr)
            case '7':
                exit()
            case _:
                 print("Invalid choice. Please enter a number between 1 and 7.")

if __name__ == "__main__":
    main()
```
